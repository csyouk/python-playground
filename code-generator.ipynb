```python
from random import randint


class ID:
    ids = []
    
    @classmethod
    def create_id(cls, minimum, maximum):
        digits = []
        number = randint(minimum, maximum)
        while number:
            digit = number%10
            digits.append(digit)
            number = number - digit
            number /=10
            number = int(number)
            if not number:
                if len(set(digits)) != len(str(minimum)):
                    return ID.create_id(minimum, maximum)
                else:
                    new = int(''.join(map(str, digits[::-1])))
                    ID.ids.append(new)
                    return new
                
    @classmethod
    def check_duplicate_id(cls, new):
        if new in generated_ids():
            new = create_id()
            return check_duplicate_id(new)
        else:
            return new

    @classmethod
    def generated_ids(cls):
        return ID.ids
    
    @classmethod
    def flush_ids(cls):
        ID.ids = []

    @classmethod
    def create_codes(cls, count, minimum, maximum):
        for _ in range(count):
            ID.create_id(minimum, maximum)
    
    @classmethod
    def check_three_quarter_pool(cls, minimum, maximum, linear_list):
        ranges = maximum - (minimum - 1)
        quarted = ranges / 3
        """
            linear_list에 들어있는 code set들의 분포에 대해서 알아보고 싶다. 
            quarter 마다 몇 개의 code들이 있는지 알아낸다.
        """
        counter = [0]*3
        for element in linear_list:
            if element < minimum + quarted:
                counter[0] += 1
            elif (minimum + quarted) <= element < (minimum + quarted*2):
                counter[1] += 1
            else:
                counter[2] += 1
        print(counter)
```
